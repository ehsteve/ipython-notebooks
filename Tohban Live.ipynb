```python
import urllib2
```

```python
spacecraft_status_url = 'http://soleil.ssl.berkeley.edu/ground_systems/rhessi_observatory_status.html'
response = urllib2.urlopen(spacecraft_status_url)
opener = urllib2.build_opener()
opener.addheaders = [('User-agent', 'Mozilla/5.0')]
response = opener.open(spacecraft_status_url)
```

```python
from bs4 import BeautifulSoup
data = response.read()
data = data.decode('UTF-8', 'ignore')
data = data.encode('UTF-8', 'ignore')
soup = BeautifulSoup(data)
print u""+soup.prettify('UTF-8').decode('UTF-8')
```
