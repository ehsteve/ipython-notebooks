```python
def create_cdict(r, g, b):
    """ Create the color tuples in the correct format"""
    i = np.linspace(0, 1, r0.size)
    
    cdict = dict(
        (name, list(zip(i, el/255.0, el/255.0)))
        for el, name in [(r, 'red'),  (g, 'green'), (b, 'blue')]
    )
    return cdict
```
