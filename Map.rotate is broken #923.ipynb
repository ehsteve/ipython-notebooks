```python
import sunpy
from sunpy.map import Map
from matplotlib import pyplot as plt
```

```python
file = '/Users/schriste/Downloads/hmi.Ic_45s_nrt.20130514_020000_TAI.2.continuum.fits'
```

```python
hmi_map = Map(file)
```

```python
plt.figure()
hmi_map.plot()
hmi_map.draw_limb()
plt.show()
```

```python
print(hmi_map.rsun_arcseconds)
```

```python
plt.figure()
plt.subplot(221)
smap = hmi_map.submap([925, 975], [-50, 50])
smap.plot()
smap.draw_limb()
plt.subplot(222)
smap = hmi_map.submap([-975, -925], [-50, 50])
smap.plot()
smap.draw_limb()
plt.subplot(223)
smap = hmi_map.submap([-50, 50], [-975, -925])
smap.plot()
smap.draw_limb()
plt.subplot(224)
smap = hmi_map.submap([-50, 50], [925, 975])
smap.plot()
smap.draw_limb()
plt.show()
```

now rotate the map

```python
print(hmi_map.data_to_pixel(0, 'x'), hmi_map.data_to_pixel(0, 'y'))
```

```python
print(hmi_map.rotation_angle)
hmi_rmap = hmi_map.rotate(angle=hmi_map.rotation_angle['y'], rotation_center=[hmi_map.data_to_pixel(0, 'y'),hmi_map.data_to_pixel(0, 'x')])
```

```python
plt.figure()
plt.subplot(221)
srmap = hmi_rmap.submap([925, 975], [-50, 50])
print(srmap.data_to_pixel(0, 'x'), srmap.data_to_pixel(0, 'y'))
srmap.plot()
srmap.draw_limb()
plt.subplot(222)
srmap = hmi_rmap.submap([-975, -925], [-50, 50])
srmap.plot()
srmap.draw_limb()
plt.subplot(223)
srmap = hmi_rmap.submap( [-50, 50], [-975, -925])
srmap.plot()
srmap.draw_limb()
plt.subplot(224)
srmap = hmi_rmap.submap([-50, 50], [925, 975])
srmap.plot()
srmap.draw_limb()
plt.show()
```
