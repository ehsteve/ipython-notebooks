```python
fe14 = ch.ion('fe_14', temperature = 2.e+6, density = 1.e+9)
wvl = wvl=200. + 0.125*arange(801)
fe14.spectrum(wvl)
plt.plot(wvl, fe14.Spectrum['intensity'])
```

```python
temperature = [1.e+6, 2.e+6]
density = 1.e+9
wvl = 200. + 0.05*arange(2001)
emeasure = [1.e+25 ,1.e+25]
s = ch.spectrum(temperature, density, wvl, filter = (chfilters.gaussian,.2), em = emeasure, verbose=0)
plt.subplot(311)
plt.plot(wvl, s.Spectrum['integrated'])
plt.subplot(312)
plt.plot(wvl, s.Spectrum['intensity'][0]*emeasure[0])
plt.subplot(313)
plt.plot(wvl, s.Spectrum['intensity'][1]*emeasure[1])
```
